```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Sales': [200, 250, 300, 350, 400],
    'Profit': [20, 30, 25, 35, 45],
    'Expenses': [180, 220, 275, 300, 350]
}
df = pd.DataFrame(data)

# Set color palette
sns.set_palette("muted")

# 1. Bar Chart for Sales
plt.figure(figsize=(8, 5))
plt.bar(df['Year'], df['Sales'], color='skyblue', edgecolor='black', label='Sales')
plt.xlabel('Year')
plt.ylabel('Sales (in units)')
plt.title('Annual Sales')
plt.legend()
plt.grid(True)
plt.show()

# 2. Line Chart for Profit
plt.figure(figsize=(8, 5))
plt.plot(df['Year'], df['Profit'], color='green', marker='o', linestyle='--', linewidth=2, label='Profit')
plt.xlabel('Year')
plt.ylabel('Profit (in units)')
plt.title('Annual Profit')
plt.legend()
plt.grid(True)
plt.show()

# 3. Combined Line and Bar Chart
fig, ax1 = plt.subplots(figsize=(8, 5))
ax1.bar(df['Year'], df['Sales'], color='lightcoral', label='Sales')
ax2 = ax1.twinx()
ax2.plot(df['Year'], df['Expenses'], color='blue', marker='s', label='Expenses')

ax1.set_xlabel('Year')
ax1.set_ylabel('Sales (in units)', color='lightcoral')
ax2.set_ylabel('Expenses (in units)', color='blue')
plt.title('Sales vs. Expenses Over Time')
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")
plt.show()

# 4. Scatter Plot for Profit vs. Expenses
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Expenses'], y=df['Profit'], hue=df['Year'], palette="viridis", s=100)
plt.xlabel('Expenses (in units)')
plt.ylabel('Profit (in units)')
plt.title('Profit vs. Expenses')
plt.legend(title="Year")
plt.grid(True)
plt.show()

# 5. Box Plot for Sales and Expenses distribution
plt.figure(figsize=(8, 5))
sns.boxplot(data=[df['Sales'], df['Expenses']], palette="Set2")
plt.xticks([0, 1], ['Sales', 'Expenses'])
plt.title('Distribution of Sales and Expenses')
plt.ylabel('Value (in units)')
plt.show()
```
