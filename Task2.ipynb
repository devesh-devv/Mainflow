```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('data.csv')

# Display basic information and the first few rows
print("Dataset Info:")
print(data.info())
print("\nFirst few rows of the dataset:")
print(data.head())

# Generate summary statistics
print("\nSummary Statistics:")
print(data.describe())

# Check for missing values
print("\nMissing Values in Each Column:")
print(data.isnull().sum())

# Visualizations
# 1. Distribution of Views
plt.figure(figsize=(10, 6))
sns.histplot(data['views'], bins=30, color='skyblue', kde=True)
plt.title('Distribution of Video Views')
plt.xlabel('Views')
plt.ylabel('Frequency')
plt.show()

# 2. Scatter Plot: Views vs. Likes
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='views', y='likes', color='purple')
plt.title("Scatter Plot between Views and Likes")
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()

# 3. Scatter Plot: Views vs. Comment Count
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='views', y='comment_count', color='green')
plt.title("Scatter Plot between Views and Comment Count")
plt.xlabel('Views')
plt.ylabel('Comment Count')
plt.show()

# 4. Box Plot: Views by Category
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='category_id', y='views')
plt.title("Box Plot of Views by Category ID")
plt.xlabel('Category ID')
plt.ylabel('Views')
plt.show()
```
